# Visualizations

### Lattice:  
A lattice is a discrete subgroup of $R^n$ which is constructed from all linear integer combinations of n basis vectors. Formally: if $v_i$ are basis vectors for $i \in \mathbb{N}$, then a lattice is a set of points where each point is defined as:
$$\mathcal{L} = \sum_{i=1}^{n}{a_i v_i} $$ where $a_i \in \mathbb{Z}$.  

As a concrete example, let's take two linearly independent vectors as our basis for the subspace $R^2$, $[0, 1]^t$ and $[1, 1]^t$. Then if we visualize all linear combinations of these vectors for a range of values, it would look like the following:


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pylab import rcParams
rcParams['figure.figsize'] = 10, 15
%matplotlib inline
v1, v2 = np.array([1, 0]), np.array([1, 1])

v = 15
plt.figure(figsize=(10,10));
plt.grid()

for i in range(-10, 10, 2):
    for j in range(-10, 10, 2):        
        newx = (i * v1 + j * v2)[0]
        newy = (i * v1 + j * v2)[1]
        plt.plot([0, newx], [0, newy], '-o')

plt.xticks(np.arange(-v - 5, v + 5, 1.0));
plt.yticks(np.arange(-v + 5, v - 5, 1.0));
```

### Elliptic Curve

```python
x = np.arange(-2.0, 3.0, 0.1);
y1 = np.array(np.sqrt(x*x*x - x + 6) % 79);
y2 = [-k for k in y1];
plt.plot(x, y1, 'k');
plt.plot(x, y2, 'k');
plt.grid(b=None, which='both', axis='both');
```
